# Step 1: Data Cleaning
Clean the raw ecommerce orders dataset — handle missing values, fix data types, and remove duplicates.

```python
import pandas as pd
import numpy as np

# Load raw data
df = pd.read_csv('../ecommerce_orders.csv')
print('Shape:', df.shape)
df.head()
```

```python
# Check missing values
print('Missing values per column:')
print(df.isnull().sum())
print()
print('Data types:')
print(df.dtypes)
```

```python
# Fix data types
df['order_date'] = pd.to_datetime(df['order_date'])
df['customer_age'] = df['customer_age'].fillna(df['customer_age'].median())

# Fill missing payment method with mode
df['payment_method'] = df['payment_method'].fillna(df['payment_method'].mode()[0])

print('After fixing - missing values:')
print(df.isnull().sum())
```

```python
# Remove duplicate rows
before = len(df)
df = df.drop_duplicates()
print(f'Removed {before - len(df)} duplicates. Rows remaining: {len(df)}')
```

```python
# Add useful time columns
df['month'] = df['order_date'].dt.month
df['year']  = df['order_date'].dt.year
df['day_of_week'] = df['order_date'].dt.day_name()

# Age group bins
df['age_group'] = pd.cut(df['customer_age'],
                         bins=[0, 25, 35, 45, 60, 100],
                         labels=['18-25', '26-35', '36-45', '46-60', '60+'])

print('New columns added. Sample:')
df[['order_date','month','year','day_of_week','age_group']].head()
```

```python
# Save cleaned data
df.to_csv('../ecommerce_orders_cleaned.csv', index=False)
print('Cleaned data saved to ecommerce_orders_cleaned.csv')
print('Final shape:', df.shape)
```
